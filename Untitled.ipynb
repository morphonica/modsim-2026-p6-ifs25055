```python
import random
import pandas as pd

# =========================
# PARAMETER
# =========================
N = 30  # jumlah mahasiswa
min_time = 1
max_time = 3

random.seed(42)

# =========================
# SIMULASI
# =========================
arrival_time = 0
current_time = 0

data = []

for i in range(1, N+1):
    service_time = random.uniform(min_time, max_time)

    start_time = current_time
    end_time = start_time + service_time
    wait_time = start_time - arrival_time

    data.append({
        "Mahasiswa": i,
        "Service Time": service_time,
        "Start": start_time,
        "End": end_time,
        "Waiting Time": wait_time
    })

    current_time = end_time

df = pd.DataFrame(data)

# =========================
# OUTPUT
# =========================
total_time = df["End"].iloc[-1]
avg_wait = df["Waiting Time"].mean()
utilization = (df["Service Time"].sum() / total_time) * 100

print("Total Waktu:", round(total_time,2))
print("Rata-rata Waktu Tunggu:", round(avg_wait,2))
print("Utilisasi:", round(utilization,2), "%")

df.head()
```

```python
# =========================
# VALIDASI
# =========================
expected_service_time = (min_time + max_time) / 2
simulated_avg_service = df["Service Time"].mean()

error_percent = abs(simulated_avg_service - expected_service_time) / expected_service_time * 100
accuracy = 100 - error_percent

print("\n=== VALIDASI ===")
print("Rata-rata Service (Simulasi):", round(simulated_avg_service,2))
print("Rata-rata Service (Teori):", expected_service_time)
print("Error (%):", round(error_percent,2))
print("Akurasi (%):", round(accuracy,2))
```
